```python
import pandas as pd
import requests
import lxml.html
import re
from tqdm import tqdm
from geopy.geocoders import Nominatim
from bs4 import BeautifulSoup 
import json
```

```python
headers = {"User-agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36 OPR/82.0.4227.50'}
r = requests.get('https://www.bodzio.pl/pl/salony/', headers=headers)
print(r.text)
```

```python
from bs4 import BeautifulSoup 
html = BeautifulSoup(r.text, 'lxml')
html
```

```python
lista_sklepow = html.find('div',{'class':'row departments-container'})
lista_sklepow = html.find('div',{'class':'px-3 ml-xl-0 mt-3'})
lista_sklepow
```

```python
sklepy = lista_sklepow.text
l = re.findall(r'\s\d*\.', lista_sklepow.text)
for x in l:
    sklepy = sklepy.replace(x, 'XXX')
    
sklepy = sklepy.split('XXX')
sklepy
```

```python
for x in range(len(sklepy)):
    sklepy[x] = sklepy[x].strip()
sklepy
    
```

```python
sklepy[0] = 'Andrychów, Krakowska 140'
```

```python
data = {'nazwa':[],
        'brand':[],
        'pelny_adres':[],
       'ulica':[],
       'kod_pocztowy':[],
       'miasto':[],
       'lat':[],
       'lng':[]}

for x in sklepy:
    
    miasto = x.split(',')[0]
    ulica = x.split(',')[1]
    
    data['nazwa'].append('Bodzio')
    data['brand'].append('Bodzio')
    data['ulica'].append(ulica)
    data['miasto'].append(miasto)
    data['kod_pocztowy'].append('x')
    data['lat'].append(str('x'))
    data['lng'].append(str('x'))
    data['pelny_adres'].append(str('x'))
    
df = pd.DataFrame(data)
df
```

```python
geolocator = Nominatim(user_agent="myGeocoder")
zle=[]
for index, row in tqdm(df.iterrows(), total=df.shape[0]):
    location = geolocator.geocode(row['ulica']+' '+row['miasto'])
    juz = False
    
    try:
        df.at[index, 'lat'] = location.raw['lat']
    except:
        df.at[index, 'lat'] = '-'
        zle.append(row)
        juz = True
        
    try:
        df.at[index, 'lng'] = location.raw['lon']
    except:
        df.at[index, 'lng'] = '-'
        if juz ==False:
            zle.append(row)
            juz = True
        
    try:
        df.at[index, 'kod_pocztowy'] = re.search(r'\d{2}\-\d{3}',location.raw['display_name'])[0]
    except:
        df.at[index, 'kod_pocztowy'] = '-'
        if juz ==False:
            zle.append(row)
        
print(len(zle))
df
```

```python
df.to_excel('bodzio.xlsx')
```
